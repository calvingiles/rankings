```python
mentee_ranking = {}
mentor_ranking = {}

for i, row in df3.iterrows():
    name = row["What is your first name?"].iloc[0]
    yes_mentee = row["Are you a mentor or mentee?"].iloc[1]
    yes_mentor = row["Are you a mentor or mentee?"].iloc[0]

    if (yes_mentor == "Mentor"):
        #print ("I am a mentor",name,"\n")
        my_mentees = row["Rank your choices of mentor from 1 ( most preferable) to least"]
        my_sorted_mentees = my_mentees.sort_values(ascending=True)
        mentee_ranking[name] = [
            mentee_name 
            for mentee_name, rank 
            in my_sorted_mentees.iteritems()
        ]
        #print("my mentees:",i,"\n",my_mentees,"\n",mentee_ranking)
    elif (yes_mentee == "Mentee"):
        #print("I am a mentee",name)
        my_mentors = row["Rank your choices of mentee from 1 ( most preferable) to least"]
        my_sorted_mentors = my_mentors.sort_values(ascending=True)
        mentor_ranking[name] = [
            mentor_name 
            for mentor_name, rank 
            in my_sorted_mentors.iteritems()
        ]
        #print("my mentors:", i, "\n", my_mentors, "\n", mentor_ranking)
    else:
        print("error in spreadsheet")
    
        
   
#    print (mentee_ranking)
#    print(mentor_ranking)
```
